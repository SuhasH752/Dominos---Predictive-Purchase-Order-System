```python
# Create time series version (with date as index)
df_daily_ts = df_daily.set_index('order_date').sort_index()

# Add time features
df_daily_ts['dayofweek'] = df_daily_ts.index.dayofweek
df_daily_ts['dayofmonth'] = df_daily_ts.index.day
df_daily_ts['month'] = df_daily_ts.index.month
df_daily_ts['quarter'] = df_daily_ts.index.quarter
df_daily_ts['is_weekend'] = (df_daily_ts['dayofweek'] >= 5).astype(int)

# Add lag features
df_daily_ts['revenue_lag1'] = df_daily_ts['daily_revenue'].shift(1)
df_daily_ts['revenue_lag7'] = df_daily_ts['daily_revenue'].shift(7)
df_daily_ts['revenue_lag30'] = df_daily_ts['daily_revenue'].shift(30)

# Add rolling statistics
df_daily_ts['revenue_rolling_mean_7'] = df_daily_ts['daily_revenue'].rolling(7).mean()
df_daily_ts['revenue_rolling_std_7'] = df_daily_ts['daily_revenue'].rolling(7).std()

# Clean data
df_daily_clean = df_daily_ts.dropna()

print("Time series data ready for forecasting:")
print(f"Shape: {df_daily_clean.shape}")
print(df_daily_clean[['daily_revenue', 'dayofweek', 'revenue_lag1', 'revenue_rolling_mean_7']].head())
```

```python
# Prepare features and target for modeling
X = df_daily_clean.drop('daily_revenue', axis=1)
y = df_daily_clean['daily_revenue']

# Split data chronologically
split_index = int(len(X) * 0.8)
X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

# Simple parameter optimization
improved_model = xgb.XGBRegressor(
    n_estimators=200,      # More trees
    learning_rate=0.05,    # Slower learning
    max_depth=8,           # Deeper trees
    subsample=0.8,         # Prevent overfitting
    colsample_bytree=0.8,  # Feature sampling
    random_state=42
)

improved_model.fit(X_train, y_train)
improved_pred = improved_model.predict(X_test)
improved_mae = mean_absolute_error(y_test, improved_pred)
print(f"Improved XGBoost MAE: ${improved_mae:.2f}")
```

```python
# Train final XGBoost model
final_model = xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=6, random_state=42)
final_model.fit(X_train, y_train)

# Generate 3-month forecasts
future_dates = pd.date_range(start=df_daily_clean.index.max() + pd.Timedelta(days=1), periods=90, freq='D')
future_data = []

for i, date in enumerate(future_dates):
    features = {}
    
    # Time features
    features['dayofweek'] = date.dayofweek
    features['dayofmonth'] = date.day
    features['month'] = date.month
    features['quarter'] = date.quarter
    features['is_weekend'] = 1 if date.dayofweek >= 5 else 0
    
    # Use recent data
    features['daily_quantity'] = df_daily_clean['daily_quantity'].iloc[-1]
    features['daily_orders'] = df_daily_clean['daily_orders'].iloc[-1]
    features['daily_ingredients_grams'] = df_daily_clean['daily_ingredients_grams'].iloc[-1]
    
    # Lag features
    if i == 0:
        features['revenue_lag1'] = df_daily_clean['daily_revenue'].iloc[-1]
        features['revenue_lag7'] = df_daily_clean['daily_revenue'].iloc[-7]
        features['revenue_lag30'] = df_daily_clean['daily_revenue'].iloc[-30]
    else:
        features['revenue_lag1'] = 0
        features['revenue_lag7'] = 0
        features['revenue_lag30'] = 0
    
    # Rolling stats
    recent_data = df_daily_clean['daily_revenue'].tail(7)
    features['revenue_rolling_mean_7'] = recent_data.mean()
    features['revenue_rolling_std_7'] = recent_data.std()
    
    future_data.append(features)

future_df = pd.DataFrame(future_data, index=future_dates)
future_df = future_df[X.columns]
future_predictions = final_model.predict(future_df)

# Create final forecast
forecast_3months = pd.DataFrame({
    'date': future_dates,
    'predicted_revenue': future_predictions
})

print("3-MONTH SALES FORECAST:")
print(forecast_3months.head(10))
```

```python
# Calculate ingredient requirements
total_predicted_revenue = forecast_3months['predicted_revenue'].sum()
total_predicted_pizzas = total_predicted_revenue / df_merged['unit_price'].mean()

# Get pizza distribution
pizza_distribution = df_merged.groupby('pizza_name_id')['quantity'].sum() / df_merged['quantity'].sum()

# Calculate predicted quantities for each pizza
predicted_quantities = (pizza_distribution * total_predicted_pizzas).round().reset_index()
predicted_quantities.columns = ['pizza_name_id', 'predicted_quantity']

# Merge with ingredient data
ingredient_details = df_ingredients.groupby(['pizza_name_id', 'pizza_ingredients'])['Items_Qty_In_Grams'].mean().reset_index()
ingredient_requirements = ingredient_details.merge(predicted_quantities, on='pizza_name_id')

# Calculate total ingredients needed
ingredient_requirements['total_kg'] = (ingredient_requirements['Items_Qty_In_Grams'] * ingredient_requirements['predicted_quantity']) / 1000

# Final purchase order
purchase_order = ingredient_requirements.groupby('pizza_ingredients')['total_kg'].sum().sort_values(ascending=False)

print("="*60)
print("FINAL PURCHASE ORDER - NEXT 3 MONTHS")
print("="*60)
print(f"Total predicted pizzas: {total_predicted_pizzas:.0f}")
print(f"Total predicted revenue: ${total_predicted_revenue:,.0f}")
print(f"Total ingredients needed: {purchase_order.sum():.0f} kg")
print("\nTOP INGREDIENTS REQUIRED:")
print(purchase_order.round(1))
```

```python
# Calculate predicted quantities for each pizza type
pizza_distribution = df_merged.groupby(['pizza_name_id', 'pizza_name'])['quantity'].sum() / df_merged['quantity'].sum()
pizza_distribution = pizza_distribution.reset_index()
pizza_distribution.columns = ['pizza_name_id', 'pizza_name', 'sales_percentage']

# Calculate predicted quantities
pizza_distribution['predicted_quantity'] = (pizza_distribution['sales_percentage'] * total_predicted_pizzas).round().astype(int)

# Sort by predicted sales
pizza_forecast = pizza_distribution[['pizza_name', 'predicted_quantity', 'sales_percentage']].sort_values('predicted_quantity', ascending=False)

print("="*60)
print("PIZZA SALES FORECAST - NEXT 3 MONTHS")
print("="*60)
print(f"Total pizzas predicted: {total_predicted_pizzas:.0f}")
print("\nTOP 20 PIZZAS BY PREDICTED DEMAND:")
print(pizza_forecast.head(20))

# Show pizza categories performance
pizza_with_categories = df_merged[['pizza_name', 'pizza_category']].drop_duplicates()
category_forecast = pizza_forecast.merge(pizza_with_categories, on='pizza_name', how='left')

category_summary = category_forecast.groupby('pizza_category')['predicted_quantity'].sum().sort_values(ascending=False)

print("\n" + "="*40)
print("PIZZA CATEGORY FORECAST")
print("="*40)
for category, quantity in category_summary.items():
    percentage = (quantity / total_predicted_pizzas) * 100
    print(f"{category}: {quantity:.0f} pizzas ({percentage:.1f}%)")
```

```python
# Merge with ingredient data for cost analysis
pizza_production_plan = pizza_forecast.merge(
    df_ingredients.groupby('pizza_name')['Items_Qty_In_Grams'].sum().reset_index(),
    on='pizza_name',
    how='left'
)

pizza_production_plan['total_ingredients_kg'] = (pizza_production_plan['Items_Qty_In_Grams'] * pizza_production_plan['predicted_quantity']) / 1000

print("\n" + "="*50)
print("DETAILED PIZZA PRODUCTION PLAN")
print("="*50)
print("Top 15 Pizzas by Production Volume:")
production_summary = pizza_production_plan[['pizza_name', 'predicted_quantity', 'total_ingredients_kg']].sort_values('predicted_quantity', ascending=False)
print(production_summary.head(15).round(1))
```
